```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread('../../DATA/sammy_face.jpg')
#img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
#gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(img)
```

```python
edges = cv2.Canny(image=img,threshold1=0,threshold2=255)
```

```python
plt.imshow(edges)
```

```python
# Calculate a midium pixel value
med_val = np.median(img)
med_val 
```

```python
#LOWER THRESHOLD TO EITHER 0 OR 70% OF THE MEDIAN VALUE WHICHEVER IS GREATER
lower = int(max(0,0.7*med_val))
#UPPER THRESHOLD TO EITHER 130% OF THE MEDIAN OR THE MAX 255, WHICHEVER IS SMALLER
upper = int(max(83,1.3*med_val))
```

```python
lower
```

```python
upper
```

```python
edges = cv2.Canny(image=img,threshold1=lower,threshold2=upper+100)
plt.imshow(edges)
```

```python
blurred_img = cv2.blur(img,ksize=(7,7))
edges = cv2.Canny(image=blurred_img,threshold1=lower,threshold2=upper)
plt.imshow(edges)
```
