**TASK: Plot the color histograms for the RED, BLUE, and GREEN channel of the giaraffe image. Pay careful attention to the ordering of the channels.**

```python
giraffes = cv2.imread('../../DATA/giraffes.jpg')
giraffes = cv2.cvtColor(giraffes,cv2.COLOR_BGR2RGB)
```

```python

def showHistogram(pic,title):
    img=pic
    color = ('b','g','r')
    for i, color in enumerate(color):
        histr =cv2.calcHist([img], [i], None, [256],[0,256])
        plt.plot(histr,color=color)
        plt.xlim([0,256])
        #plt.ylim([0,500000]) #

    plt.title('HISTOGRAM FOR '+title)
```

```python
showHistogram(giraffes, 'GIRAFES')
```
