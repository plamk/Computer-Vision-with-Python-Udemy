```python
img1 = cv2.imread('../../DATA/dog_backpack.png')
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
```

```python
#First, we have to resize the images so we can use addWeighted() when the size of images is different
img1= cv2.resize(img1, (1200,1200))
img2= cv2.resize(img2, (1200,1200))
```

```python
#BLENDING IMAGES OF THE SAME SIZE (we can get the error because the images have different sizes)
blended = cv2.addWeighted(img1, alpha=0.8, src2=img2, beta=0.5, gamma=0)
```

```python
plt.imshow(blended)
```

```python
img1 = cv2.imread('../../DATA/dog_backpack.png')
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
img2 = cv2.imread('../../DATA/watermark_no_copy.png')
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
```

```python
img2 = cv2.resize(img2,(600,600))
```

```python
plt.imshow(img2)
```

```python
large_img = img1
small_img = img2
```

```python
x_offset = 0
y_offset = 0
```

```python
small_img.shape
```

```python
x_end = x_offset + small_img.shape[1]
y_end = y_offset + small_img.shape[0]
```

```python
large_img[y_offset:y_end,x_offset:x_end] = small_img
plt.imshow(large_img)
```
