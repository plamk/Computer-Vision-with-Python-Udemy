```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
```

```python
img = cv2.imread('../../DATA/00-puppy.jpg')
```

```python
plt.imshow(img)
```

```python
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
imgHSV = np.copy(img)
imgHSV = cv2.cvtColor(imgHSV,cv2.COLOR_RGB2HSV)
plt.imshow(imgHSV)
```

```python
imgHLS = np.copy(img)
imgHLS = cv2.cvtColor(imgHLS,cv2.COLOR_RGB2HLS)
plt.imshow(imgHLS)
```
