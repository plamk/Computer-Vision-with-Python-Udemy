```python
#Function to calculate histograms blue values of picture
def showHistoB(pic):
    h = cv2.calcHist([pic], channels=[0], mask=None,histSize=[256],ranges=[0,256])
    return plt.plot(h)
```

```python
#Function to calculate histograms red values of picture
def showHistoR(pic):
    h = cv2.calcHist([pic], channels=[2], mask=None,histSize=[256],ranges=[0,256])
    return plt.plot(h)
```

```python
#My function 
def showHistogram(pic,title):
    img=pic
    color = ('b','g','r')
    for i, color in enumerate(color):
        histr =cv2.calcHist([img], [i], None, [256],[0,256])
        plt.plot(histr,color=color)
        plt.xlim([0,256])
        #plt.ylim([0,500000]) #

    plt.title('HISTOGRAM FOR '+title)
```

```python
showHistogram(dark_horse,'DARK HORSE')
```

```python
showHistogram(blue_bricks,'BLUE BRICKS')
```

```python
showHistogram(rainbow,'RAINBOW')
```
